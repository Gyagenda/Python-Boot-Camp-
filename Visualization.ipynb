# using matplotlib 

```python

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
fig1=plt.figure()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()


fig1.savefig('myfigure.png')
```

```python
fig2=plt.figure()
plt.plot(x,np.sin(x))
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("y")
```

# plot a histogram 

```python
data=np.random.randn(1000)
fig3=plt.figure()
plt.hist(data)
plt.hist(data,bins=20)
fig3.savefig("myhist.png")
```

```python
norm=np.random.standard_normal(1000)
fig4=plt.figure()
plt.hist(norm,bins=50)
fig4.savefig('normal.png')
plt.title("Normal Distrubtion")
plt.xlabel("x")
plt.ylabel("y")
```
